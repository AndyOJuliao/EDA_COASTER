```python
df_corr = df[['Year_Introduced', 'Speed_mph', 'Gforce_Clean', 'Inversions_Clean']].dropna().corr()

#This section shows us how the features are correlated, by all means, it show the statistical goal between -1 and 1 of 
#correlatiion

df_corr
```

```python
sns.heatmap(df_corr, annot = True)

#in the line above, we create the df_corr wich basically is the numerical correlation between these features, here, we're using
# the heatmap to put into a visualization what the correlation is about. Using the notation annot=True to show these values.
```

# 5 Asking questions about the data 

```python
df.head() #What are the locations with the fastest roller coaster (min of 10 coaster of that location) ?
```

```python
df['Location'].value_counts()
```

```python
ax = df.query('Location != "Other"').groupby('Location')['Speed_mph'].agg(['mean', 'count']).query('count >= 10 ')\
    .sort_values('mean')['mean']\
    .plot(kind = 'barh', figsize=(12,5) , title = 'Average Coaster Speed by Location')
          
ax.set_xlabel('Average Coaster Speed')
ax.set_ylabel('Locations')


#Here we are using a complex code line to set: A query into the df where location is not equal to other, and using the
#function groupby to group the location by the means of Speed mph, and finally using the agg function to agregate those
#informations by the mean value and the count of them
```
